# Calculating Probability Of Winning Next Point (Given Previous Point Won)

### Dictionary of all the games Alcaraz played and his opponent

```python
a_games = {}
players = set()
for r in np_alcaraz:
    players.add((r['player1'], r['player2']))
    if r['player1'] == "Carlos Alcaraz":
        a_games[r['match_id']] = r['player2']
    else:
        a_games[r['match_id']] = r['player1']

print(a_games)
print(len(a_games))
print(players)
```

```python
matches = {}

for m in a_games:
    matches[m] = list()

for m in a_games:
    for r in np_alcaraz:
        if r['match_id'] == m:
            matches[m].append(r)

print(matches['2023-wimbledon-1301'])
```
