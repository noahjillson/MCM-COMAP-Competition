```python
import csv
# Specify the path to your CSV file
csv_file_path = '/Users/noahjillson/Desktop/MCM/Wimbledon_featured_matches.csv'

# Open the CSV file in read mode
alcaraz_interactions = []
with open(csv_file_path, 'r') as file:
    # Create a CSV reader object that returns rows as dictionaries
    csv_reader = csv.DictReader(file)

    # Iterate through the rows in the CSV file
    for row in csv_reader:
        # Each row is a dictionary with keys as header names
        if (row['player1'] == "Carlos Alcaraz" or row['player2'] == "Carlos Alcaraz"):
            alcaraz_interactions.append(row)
```

```python
np_alcaraz = np.array(alcaraz_interactions)
```

# Calculating Probability Of Winning Next Point (Given Previous Point Won)

### Dictionary of all the games Alcaraz played and his opponent

```python
a_games = {}
players = set()
for r in np_alcaraz:
    players.add((r['player1'], r['player2']))
    if r['player1'] == "Carlos Alcaraz":
        a_games[r['match_id']] = r['player2']
    else:
        a_games[r['match_id']] = r['player1']

print(a_games)
print(len(a_games))
print(players)
```

```python
matches = {}

for m in a_games:
    matches[m] = list()

for m in a_games:
    for r in np_alcaraz:
        if r['match_id'] == m:
            matches[m].append(r)

print(matches['2023-wimbledon-1301'])
```

```python
p_sum = 0
p_consec_sum = 0
for mr in matches.values():
    pl = None
    cpnum = '2'
    if mr[0]['player1'] == 'Carlos Alcaraz':
        cpnum = '1'
    for r in mr:
        if pl is None:
            pl = r['point_victor']
            continue
        # If Calros won last point
        if cpnum == pl:
            # If Carlos won current point
            if r['point_victor'] == cpnum:
                # inc sum to indicate carlos won current point after winning the previous one
                p_consec_sum += 1 
            # Increment sum of opportunities for carlso to win the next point
            p_sum += 1

        pl = r['point_victor']

print(p_sum, p_consec_sum, p_consec_sum/p_sum)
```
